```python
import pandas as pd

path = '../data/Metro_Interstate_Traffic_Volume.csv'
data = pd.read_csv(path)
data
```

```python
data = data.sort_values(by=["temp", "rain_1h", "snow_1h", 'clouds_all'])
data.head()
```

### feature analyze

```python
data.info()
```

```python
data['holiday'] = data['holiday'].fillna('No')
data.count()
```

### Remove outlier

```python
# remove 0 temp as outliers
temp_outlier = data[data['temp'] == 0]
data = data.drop(data[data['temp'] == 0].index)
data.head()
```

```python
#  remove numeric value outliers in rain_1h
rain_outlier = data[data['rain_1h'] > 100]
data = data.drop(data[data['rain_1h'] > 100].index)
data.head()
```

### Categorize

```python
import numpy as np
# convert the rain_1h to category
data['rain_1h'] = np.where(data['rain_1h'] == 0, 'No rain', 'Have rain')
data
```

```python
# convert snow columns to category values
data['snow_1h'] = np.where(data['snow_1h'] == 0, 'No snow', 'Have snow')
data
```

```python
def categorize_cloud(column, p25, p50, p75):
  """
  Function use to categorize based on percentile
  - apply for cloud column in dataset
  """
  if isinstance(column, (int, float)):
    if column <= p25:
      column = 'Clear' 
    elif column <= p50:
      column = 'Party Cloudy'
    elif column <= p75:
      column = 'Mostly Cloudy'
    else:
      column = 'Overcast'
  
  return column
```

```python
p25 = np.percentile(data['clouds_all'], 25)
p50 = np.percentile(data['clouds_all'], 50)
p75 = np.percentile(data['clouds_all'], 75)
data['clouds_all'] = data['clouds_all'].apply(lambda x: categorize_cloud(x, p25, p50, p75))
data
```

### split datetime to detailed numeric info

```python
from datetime import datetime

def convert_date_time(col):
  col = pd.to_datetime(col, format='%d-%m-%Y %H:%M')
  df= pd.DataFrame()
  
  # split date_time column into more detailed data
  df['year'] = col.dt.year
  df['month'] = col.dt.month
  df['day'] = col.dt.day
  df['hour'] = col.dt.hour
  # from monday to sunday in ordinary
  df['day_in_week'] = col.dt.weekday
  df['is_weekend'] = col.isin([5, 6]).astype(int)
  
  return df
  
```

```python
date_time_features = convert_date_time(data['date_time'])
date_time_features = pd.DataFrame(date_time_features)
date_time_features.head()
```

```python
processed_data = pd.concat([data, date_time_features], axis=1)
processed_data = processed_data.drop(columns=['date_time'])
processed_data.head()
```

### Pre-process weather columns

```python
def group_weather(weather):
  """
  weather description & weather main process: 
  Cloud
  Clear
  Mist = mist + haze + fog + smoke
  Rain = rain + prizzle + thunderstorm + squall
  Snow
  """
  weather_map = {
        'Clouds': 'Clouds',
        'Clear': 'Clear',
        'Mist': 'Mist',
        'Haze': 'Mist',
        'Fog': 'Mist',
        'Smoke': 'Mist',
        'Rain': 'Rain',
        'Drizzle': 'Rain',
        'Thunderstorm': 'Rain',
        'Squall': 'Rain',
        'Snow': 'Snow'
    }
  simplified_weather = weather.map(weather_map)
    
  # Handle any unmapped values (optional, if needed)
  simplified_weather.fillna('Unknown', inplace=True)
  
  return simplified_weather
```

```python
processed_data['weather_main'] = group_weather(processed_data['weather_main'])
processed_data = processed_data.drop(processed_data[['weather_description']], axis=1)
```
